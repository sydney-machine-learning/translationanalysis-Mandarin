```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "http://www.coldbacon.com/writing/luxun-calltoarms.html#AhQ"
```

```python
#translator -- Yang Hsien-yi and Gladys Yang
response = requests.get(url)

# Parse the webpage content
soup = BeautifulSoup(response.content, 'html.parser')
```

```python

# Find all text elements in the page
text_elements = soup.find_all(text=True)

# chapter four
start_marker = "THE TRAGEDY OF LOVE"
end_marker = "____"
# Extract all paragraph tags
page_text = soup.get_text(separator='\n')  # Use double newlines to separate paragraphs
# Search for the start and end positions of the markers
start_pos = page_text.lower().find(start_marker.lower())
end_pos = page_text.lower().find(end_marker.lower(), start_pos)

if start_pos != -1 and end_pos != -1:
    start_pos += len(start_marker)  # Adjust to start after the marker
    # Extract the text between the markers
    extracted_content = page_text[start_pos:end_pos].strip()

    # Split the extracted content into paragraphs
    paragraphs = [p.strip() for p in extracted_content.split('\n\n') if p.strip()]

    # Create a DataFrame
    df_chapt4_hsien = pd.DataFrame(paragraphs, columns=['Paragraph'])
else:
    print("Chapter markers not found.")


df_chapt4_hsien.rename(columns={'Paragraph': 'English_Verse_chap4_hsien'}, inplace=True)




```

```python
#data merge


merged_rows_hsien_chapt4 = df_chapt4_hsien.iloc[31:33]
# Define a custom aggregation function
def agg(x):
    if x.dtype == 'object':
        return ' '.join(x)
    else:
        return x.sum()


merged_row_hsien_chapt4 = merged_rows_hsien_chapt4.agg(agg)

original_indices_hsien_chapt4 = merged_rows_hsien_chapt4.index

insert_index_hsien_chapt4 = original_indices_hsien_chapt4[0]

df_dropped = df_chapt4_hsien.drop(original_indices_hsien_chapt4)

merged_row_df_hsien_chapt4 = pd.DataFrame([merged_row_hsien_chapt4], index=[insert_index_hsien_chapt4])

df_with_merged_hsien_chapt4 = pd.concat([df_dropped.iloc[:insert_index_hsien_chapt4], merged_row_df_hsien_chapt4, df_dropped.iloc[insert_index_hsien_chapt4:]]).sort_index().reset_index(drop=True)



```

```python
#google translation
import requests
import pandas as pd
import re

# URL of the text file
url = "http://www.xys.org/xys/classics/Lu-Xun/Nahan/aq.txt"

# Send a request to the text file
response = requests.get(url)
page_text = response.text

# Define the start and end markers in Chinese
start_marker_ch = "恋爱的悲剧"  # Example start marker in Chinese (Tragedy of Love)
end_marker_ch = "第五章"  # Example end marker in Chinese (Chapter 5)

# Search for the start and end positions of the markers
start_pos_ch = page_text.find(start_marker_ch)
end_pos_ch = page_text.find(end_marker_ch, start_pos_ch)

if start_pos_ch != -1 and end_pos_ch != -1:
    start_pos_ch += len(start_marker_ch)  # Adjust to start after the marker
    # Extract the text between the markers
    extracted_content_ch = page_text[start_pos_ch:end_pos_ch].strip()

    # Split the extracted content into paragraphs based on double newlines
    paragraphs_ch = re.split(r'\n\s*\n', extracted_content_ch)
    # Replace single newlines within paragraphs with spaces
    paragraphs_ch = [re.sub(r'\s*\n\s*', ' ', p).strip() for p in paragraphs_ch if p.strip()]

    # Create a DataFrame
    df_ch = pd.DataFrame(paragraphs_ch, columns=['Paragraph'])
    print(df_ch)
else:
    print("Chapter markers not found.")

```

```python
merged_rows = df_ch.iloc[39:44]

# Define a custom aggregation function
def agg(x):
    if x.dtype == 'object':
        return ' '.join(x)
    else:
        return x.sum()


merged_row = merged_rows.agg(agg)

original_indices = merged_rows.index

insert_index = original_indices[0]

df_dropped = df_ch.drop(original_indices)

merged_row_df = pd.DataFrame([merged_row], index=[insert_index])

df_with_merged = pd.concat([df_dropped.iloc[:insert_index], merged_row_df, df_dropped.iloc[insert_index:]]).sort_index().reset_index(drop=True)

print(df_with_merged)
```

```python
from googletrans import Translator, LANGUAGES

translator = Translator()

for col in df_with_merged.columns:
    for i in range(len(df_with_merged)):
        text = df_with_merged.at[i, col]
        translated = translator.translate(text, src='zh-cn', dest='en')
        df_with_merged.at[i, col] = translated.text

df_with_merged.iloc[30]
```

```python
#Julia Lovell
import PyPDF2

# Open the PDF file in read-binary mode
with open('/Users/wangxuechun/Desktop/data_preparation/Version_by_YiyunLi_chapt6.pdf', 'rb') as pdf_file:
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Initialize an empty string to store the text
    pdf_text = ''

    # Iterate through the pages and extract text
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        pdf_text += page.extract_text()
    # test

print(pdf_text)
```

```python
import os
import pandas as pd

#create a dataframe to store the chinese paragraph and english paragraph
df_yiyun_chapt6 = pd.DataFrame({
    'English_Verse_chap6_julia': verses,
})

#store the csv format in the desktop 
desktop = os.path.join(os.path.join(os.path.expanduser('~')), 'Desktop')
filepath = os.path.join(desktop, 'chapter_6_eng_version_julia.csv')
df_yiyun_chapt6.to_csv(filepath, index=False)
```

```python
#preprocesss data
df_julia_chapt4 = pd.read_csv('/Users/wangxuechun/unsw/thesis-repo/data_processing/chapter_4_eng_version_julia.csv')
print(df_julia_chapt4)
merged_rows_julia = df_julia_chapt4.iloc[38:43]

# Define a custom aggregation function
def agg(x):
    if x.dtype == 'object':
        return ' '.join(x)
    else:
        return x.sum()


merged_row_julia = merged_rows_julia.agg(agg)

original_indices_julia = merged_rows_julia.index

insert_index_julia = original_indices_julia[0]

df_dropped = df_julia_chapt4.drop(original_indices_julia)

merged_row_df_julia = pd.DataFrame([merged_row_julia], index=[insert_index_julia])

df_with_merged_julia = pd.concat([df_dropped.iloc[:insert_index_julia], merged_row_df_julia, df_dropped.iloc[insert_index_julia:]]).sort_index().reset_index(drop=True)

print(df_with_merged_julia)

file_path_julia_chapt4 = '/Users/wangxuechun/unsw/thesis-repo/data_processing/chapter_4_eng_version_julia.csv'
df_with_merged_julia.to_csv(file_path_julia_chapt4, index=False)
```
